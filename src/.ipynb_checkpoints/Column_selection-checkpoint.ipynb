```python
import pandas as pd
import pandasql as ps
```

```python
raw_data = pd.read_csv('people_data_u19.csv')
raw_data = ps.sqldf("""SELECT * FROM raw_data""")
raw_data
```

```python
data = ps.sqldf("""SELECT time_string as `discovery date`,
                   name_key_a as `id1`,
                   name_key_b as `id2`
                 FROM raw_data
            
            UNION
            
            SELECT time_string as `discovery date`,
                   name_key_b as `id1`,
                   name_key_a as `id2`
                 FROM raw_data
            """)
data.to_csv('input.csv',index=False)

```
